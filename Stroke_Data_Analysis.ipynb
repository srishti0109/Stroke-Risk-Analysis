```python
import pandas as pd
```

```python
# Load healthcare dataset
df = pd.read_csv(r"C:\Users\shris\Downloads\archive (2)\healthcare-dataset-stroke-data.csv")
# Display first 5 rows of the dataset
df.head()
```

```python
# Check number of rows and columns
df.shape
```

```python
# Display all column names
df.columns
```

```python
# Get dataset information such as data types and missing values
df.info()
```

```python
# Check missing values in each column
df.isnull().sum()
```

```python
# Display summary statistics of BMI column
df["bmi"].describe()
```

```python
# Fill missing BMI values using median BMI
df["bmi"] = df["bmi"].fillna(df["bmi"].median())
```

```python
# Verify that missing BMI values have been removed
df.isnull().sum()
```

```python
# Count number of stroke and non-stroke patients
df["stroke"].value_counts()
```

```python
# Compare average age of stroke and non-stroke patients
df.groupby("stroke")["age"].mean()
```

```python
# Compare average glucose level of stroke and non-stroke patients
df.groupby("stroke")["avg_glucose_level"].mean()
```

```python
# Compare hypertension rate between stroke and non-stroke patients
df.groupby("stroke")["hypertension"].mean()
```

```python
# Compare heart disease rate between stroke and non-stroke patients
df.groupby("stroke")["heart_disease"].mean()
```

```python
# Count patients by smoking status
df["smoking_status"].value_counts()
```

```python
# Compare stroke cases across smoking status categories
pd.crosstab(df["smoking_status"], df["stroke"])
```

```python
# Calculate stroke rate percentage by smoking status
pd.crosstab(df["smoking_status"], df["stroke"], normalize="index") * 100
```

```python
# Compare stroke rate by gender
pd.crosstab(df["gender"], df["stroke"], normalize="index") * 100
```

```python
# Compare stroke rate by marital status
pd.crosstab(df["ever_married"], df["stroke"], normalize="index") * 100
```

```python
# Compare stroke rate by residence type
pd.crosstab(df["Residence_type"], df["stroke"], normalize="index") * 100
```

```python
# Compare stroke rate by work type
pd.crosstab(df["work_type"], df["stroke"], normalize="index") * 100
```

```python
# Compare average age across different work types
df.groupby("work_type")["age"].mean()
```

```python
# Compare average BMI between stroke and non-stroke patients
df.groupby("stroke")["bmi"].mean()
```

```python
# Create age groups and compare stroke rate by age group
df["age_group"] = pd.cut(
    df["age"],
    bins=[0,20,40,60,100],
    labels=["0-20","21-40","41-60","60+"]
)

pd.crosstab(df["age_group"], df["stroke"], normalize="index") * 100
```

```python
# Save cleaned dataset for Power BI
df.to_csv("cleaned_stroke_data.csv", index=False)
```
